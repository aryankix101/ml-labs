```python
from sklearn.datasets import make_circles
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
import tensorflow as tf
import pandas as pd 
from sklearn.preprocessing import normalize 
from tensorflow.keras import layers 
from tensorflow.keras import utils as np_utils
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten


X, y = make_circles(1000, shuffle=True, noise=None, random_state=42, factor=0.8)

plt.scatter(X[:,0], X[:,1], label = "original")
plt.legend()
plt.show()

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.50, random_state=42)

def create_model():
    model = tf.keras.Sequential()
    model.add(Dense(2))
    model.add(Dense(5, activation='relu'))
    model.add(Dense(5, activation='relu'))
    model.add(Dense(5, activation='relu'))
    model.add(Dense(5, activation='relu'))
    model.add(Dense(5, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))

    return model

model = create_model()
optimizer = keras.optimizers.Adam(lr=0.1)
model.compile(optimizer=optimizer, metrics=['accuracy'], loss="BinaryCrossentropy")
model_history = model.fit(X_train, y_train, batch_size=None, epochs=500, validation_data=(X_test, y_test))

plt.plot(model_history.history['accuracy'])
plt.plot(model_history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```
